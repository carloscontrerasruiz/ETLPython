# Conectar a base de datos MySql

```python
# importar librerias
import mysql.connector
import pandas as pd
from sqlalchemy import create_engine
from datetime import datetime

#to work with slice and not to throw any warning when those slices are being modified
pd.set_option('mode.chained_assignment', None)

conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="MysqlCarlos",
    database="sakila"
)

cursor = conn.cursor()

query="SELECT * FROM actor;"
cursor.execute(query)

data= cursor.fetchall()
#column names
print('COLUM NAMES AND CURSOR DESCRIPTION')
print(cursor.description)
columns = [col[0] for col in cursor.description]

df=pd.DataFrame(data, columns=columns)
print('DATA HEAD')
print(df.head())

cursor.close()
conn.close()
```

# Step 2 Transform data
### Filter actors first nam start with an A

```python
df_filtered = df[df['first_name'].str.startswith('A')].copy() #Creamos una copia explicta, no modificamos el original
print(df_filtered.head())
```

### Create column named fullname using first_name and last_name

```python
df_filtered['fullname']=df['first_name'] + ' ' + df['last_name']
print(df_filtered.head())
```

### Create new column name_length with the totale length of the name

```python
df_filtered['name_length'] = df_filtered['fullname'].str.len() #possible faster
#df_filtered['name_length'] = df_filtered['fullname'].apply(len)
print(df_filtered.head())
```

### Transform first_name and last_name to upper (This should be done before creating the fullname)

```python
df_filtered['first_name'] = df_filtered['first_name'].str.upper()
df_filtered['last_name'] = df_filtered['last_name'].str.upper()
df_filtered
```

### Filter actors with fullname greater than 10 characters

```python
#Both way work
#df_filtered = df_filtered[df_filtered['fullname'].str.len() > 10]
df_filtered = df_filtered[df_filtered['name_length'] > 10]
df_filtered
```

### Get Actor registry year

```python
df_filtered['registration_year']=pd.to_datetime(df_filtered['last_update']).dt.year
df_filtered
```

### Group by last_name and count how many actors are with the same

```python
df_lastname_group = df_filtered.groupby('last_name').agg(
    total_actors=('actor_id', 'count')
) #without reset_index use last_name as index

df_lastname_group
```

```python
df_lastname_group = df_filtered.groupby('last_name').agg(
    total_actors=('actor_id', 'count')
).reset_index() #with reset_index use a whole new number index as index

df_lastname_group
```

### Add column to show if actor has a unique name 

```python
df_filtered['isUniqueName']=df_filtered.groupby('fullname')['fullname'].transform('count') == 1
df_filtered
```

### Add column actor_status to show if actor has more than 10 years registered

```python
def getActorStatus(datetime):
    actorYear = pd.to_datetime(datetime).year
    currentYear = datetime.now().year
    if(currentYear-actorYear > 10):
        return 'SENIOR'
    return 'ROOKIE'

#Fails because The truth value of a Series is ambiguous. Use a.empty, a.bool(), a.item(), a.any() or a.all()
#df_filtered['actor_status'] = 'SENIOR' if (datetime.now().year - pd.to_datetime(df['last_update']).dt.year) > 10  else 'ROOKIE'
df_filtered['actor_status'] = df_filtered['last_update'].apply(getActorStatus)
df_filtered
```

### Delete last_update column

```python
df_filtered=df_filtered.drop('last_update', axis=1)
df_filtered
```

## Create database and tables to dump filtered|transform info

```python
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="MysqlCarlos"
)

cursor = conn.cursor()

createDatabaseQuery= "CREATE DATABASE IF NOT EXISTS sakila_etl"
cursor.execute(createDatabaseQuery)
cursor.execute("USE sakila_etl")

create_actor_transformed_table = """
CREATE TABLE IF NOT EXISTS actor_transformed (
    actor_id INT PRIMARY KEY,              -- Identificador nico del actor
    first_name VARCHAR(50),                 -- Nombre del actor
    last_name VARCHAR(50),                  -- Apellido del actor
    full_name VARCHAR(100),                 -- Nombre completo concatenado
    name_length INT,                        -- Longitud del nombre completo
    registration_year INT,                  -- Ao de registro (basado en 'last_update')
    is_unique_name BOOLEAN,                 -- Si el nombre es nico en la base de datos
    actor_status VARCHAR(10)                -- Estado del actor (Activo/Inactivo)
);
"""

create_actor_count_by_last_name_table = """
CREATE TABLE IF NOT EXISTS actor_count_by_last_name (
    last_name VARCHAR(50) PRIMARY KEY,      -- Apellido del actor
    total_actors INT                        -- Total de actores con ese apellido
);
"""

cursor.execute(create_actor_transformed_table)
cursor.execute(create_actor_count_by_last_name_table)

conn.commit()
cursor.close()
conn.close()
```
